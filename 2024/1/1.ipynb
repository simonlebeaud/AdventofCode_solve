```python
import numpy as np
```

```python
f = open("inputs.txt", "r")
first = []
second = []
result0 = 0
result1 = 0
```

```python
for l in f:
    first.append(l.split("   ")[0])
    second.append(l.split("   ")[1])
print(second)
```

```python
first = sorted(list(map(int,first)))
second = sorted(list(map(int, second)))
```

```python
for i in range(len(first)):
    result1 += first[i]*second.count(first[i])
```

```python
for i in range(len(first)):
    m = first.pop(0)
    ms = second.pop(0)
    result0 += abs(m-ms)
```

```python
print(result0, result1)
```
