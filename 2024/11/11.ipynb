```python

def blinck_25(rocks):
    def split_rock(rocks):
        rx = []

        for rock in rocks:
            if rock=='0':
                rx.append('1')
                continue
            
            l = len(rock)
            if l%2==0:
                mid=l//2
                rx.append(rock[:mid])
                rx.append('0' if rock[mid:].strip('0')=='' else rock[mid:].lstrip('0'))
            else:
                rx.append(str(int(rock)*2024))
        return rx

    rocks = rocks.split()
    for _ in range(25):
        rocks = split_rock(rocks)
    print(len(rocks))

blinck_25(rocks)
```

```python
from collections import Counter

def split_number_count(num):
    """Returns the two parts of a number if it has an even number of digits."""
    num_str = str(num)
    mid = len(num_str) // 2
    left = int(num_str[:mid])
    right = int(num_str[mid:])
    return left, right

def evolve_stones_optimized(initial_stones, blinks):
    # Count occurrences of each stone value
    stone_counts = Counter(initial_stones)
    for _ in range(blinks):
        new_counts = Counter()
        for stone, count in stone_counts.items():
            if stone == 0:
                # Rule 1: Stone with 0 becomes 1
                new_counts[1] += count
            elif len(str(stone)) % 2 == 0:
                # Rule 2: Stone with even digits splits into two parts
                left, right = split_number_count(stone)
                new_counts[left] += count
                new_counts[right] += count
            else:
                # Rule 3: Multiply by 2024
                new_counts[stone * 2024] += count

        # Update stone counts for the next blink
        stone_counts = new_counts

    # Return the total number of stones after all blinks
    return sum(stone_counts.values())

# Initial arrangement of stones (your example input)
f = open('input.txt','r')
initial_stones = []
for l in f:
    initial_stones=list(map(int,l.strip().split()))


# Part 1: Number of stones after 25 blinks
stones_after_25_blinks = evolve_stones_optimized(initial_stones, 25)
print("Number of stones after 25 blinks:", stones_after_25_blinks)

# Part 2: Number of stones after 75 blinks
stones_after_75_blinks = evolve_stones_optimized(initial_stones, 75)
print("Number of stones after 75 blinks:", stones_after_75_blinks)
```
